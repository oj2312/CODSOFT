## Handling Outliers


```python
data = df.copy()
```

```python
data.drop(columns = ["customer name" , "customer e-mail" , "country"], axis = 1 , inplace = True)
```

```python
percentile25 = []
percentile75 = []
```

```python
for i in col:
    percentile25.append(data[i].quantile(0.25))
    percentile75.append(data[i].quantile(0.75))
```

```python
print(percentile25)
print(percentile75)
```

```python
lower_limit = []
upper_limit = []
iqr = []
    
for i in range(len(col)):
        iqr.append((percentile75[i] - percentile25[i]))
        lower_limit.append(percentile25[i] - 1.5 * iqr[i])
        upper_limit.append(percentile75[i] + 1.5 * iqr[i])
```

```python
print(iqr)
print(lower_limit)
print(upper_limit)
```

```python
j = 0
for i in col:
    data[i] = np.where(
    data[i] > upper_limit[j],
    upper_limit[j] ,
    np.where(data[i] < lower_limit[j],
             lower_limit[j],
             data[i]
            )
       )
    j = j+1
```

```python
for i in col:
    sns.boxplot(data[i])
    plt.xlabel(i)
    plt.show()
    print("\n\n")
```
